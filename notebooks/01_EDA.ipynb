# Reduzir tamanho da memoria

```python
colunas_valores_inteiros = []

for coluna in df.columns:
    if df[coluna].apply(lambda x: x.is_integer()).all():
        colunas_valores_inteiros.append(coluna)

colunas_valores_inteiros
```

```python
df.info()
```

```python
df[colunas_valores_inteiros] = df[colunas_valores_inteiros].astype(int)

df.info()
```

```python
df[colunas_valores_inteiros] = df[colunas_valores_inteiros].apply(
    pd.to_numeric,
    downcast="integer"
)

df.info()
```

```python
colunas_valores_nao_inteiros = df.columns.difference(colunas_valores_inteiros)

colunas_valores_nao_inteiros 
```

```python
df[colunas_valores_nao_inteiros] = df[colunas_valores_nao_inteiros].apply(
    pd.to_numeric,
    downcast="float"
)

df.info()
```
